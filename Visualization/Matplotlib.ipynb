```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x,y)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('x vs y');           
```

```python
plt.figure(figsize = (6,2))
plt.scatter(x,y, c='#223BDC', alpha = 0.9, marker = 'X')
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('x vs y')
plt.grid()
```

```python
x = ['a','b','c','d', 'e']
y = np.random.rand(5)
```

```python
plt.figure(figsize = (4,3))
plt.bar(x,y, color = 'pink')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x vs y')
plt.grid()
```

```python
plt.figure(figsize = (4,3))
plt.barh(x,y, color = 'purple')## horizontal barplot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x vs y')
plt.grid()
```

```python
x = [3,4,5,6,7,8]
y = [4,5,6,7,8,9]
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize = (6,3))
plt.hist(data, color = 'pink')## horizontal barplot
plt.xlabel('representing unique data')
plt.ylabel('frequency of each unique value')
plt.title('Histogram plot')
plt.grid()
plt.savefig('test.png')
```

```python
x  = np.random.rand(50)
y  = np.random.rand(50)
z  = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,  alpha = 0.9)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show();
```
