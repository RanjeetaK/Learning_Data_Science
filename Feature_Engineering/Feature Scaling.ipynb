```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

### 1. Standardization

```python
import numpy as np
```

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
print(mean, std)
```

```python
normalized_data = []
for i in list(df['total_bill']):
    z_score = (i-mean)/std
    normalized_data.append(z_score)
```

```python
normalized_data
```

```python
sns.histplot(df['total_bill'], kde = True)
```

```python

sns.histplot(normalized_data, kde = True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill']])# fit calculates the mean and std for all the values
```

```python
scaler.transform(df[['total_bill']]) # transform actuallay starts scaling down the values using mean and std
```

```python
## We can aslo write as below
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```

```python
scaler.transform([[13,4]]) ## scaling for the new dataset
```

### 2. Normalization Min-Max Scaler

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['distance', 'fare', 'tip']])
```

```python
min_max.fit(df[['distance', 'fare', 'tip']])
```

```python
min_max.transform(df[['distance', 'fare', 'tip']])
```

```python
min_max.transform([[1.6, 7.0, 4]]) ## for any new datapoint
```
